```python
import numpy as np
import matplotlib.pyplot as plt

I = plt.imread("mawis.jpeg").astype(float)
plt.subplot(121)
plt.imshow(I)

P = plt.imread("mwasi_red.jpeg").astype(float)
plt.subplot(122)
plt.imshow(P)
```

```python
R = I[:,:,0]
G = I[:,:,1]
B = I[:,:,2]

Y = R + G + B

Y[Y == 0] = 100000

r = R / Y
g = G / Y
b = 1 - r - g

plt.plot(r,g,".")
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()

Y.max()
```
