```python
import numpy as np
import cv2
```

```python
vidcap = cv2.VideoCapture('pendulum_2.mov')
```

```python
success,image = vidcap.read()
count = 0
success = True
while success:
    if count < 10:
        cv2.imwrite("Pendulum_2/frame00%d.png" % count, image)
    if count >= 10 and count < 100:
        cv2.imwrite("Pendulum_2/frame0%d.png" % count, image)
    if count >= 100:
        cv2.imwrite("Pendulum_2/frame%d.png" % count, image)
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```
