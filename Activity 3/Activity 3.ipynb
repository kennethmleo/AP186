```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

### Display RGB Image / Truecolor Image

```python
RGB_image = mpimg.imread("truecolor.jpg")
RGB_image = RGB_image[2100:2800,2000:4000]

plt.figure(figsize=(20,10))
plt.imshow(RGB_image)
plt.axis('off')
plt.savefig('truecolor.jpg', bbox_inches='tight')
plt.show()
```

### Converting RGB image to Grayscale image

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
Grayscale_image = rgb2gray(RGB_image)

plt.figure(figsize=(20,10))
plt.imshow(Grayscale_image, cmap=plt.get_cmap('gray'))
plt.axis('off')
plt.savefig('grayscale.jpg', bbox_inches='tight')
plt.show()
```

### Converting RGB image to Binary image

```python
import cv2

img = cv2.imread('helium_1.jpeg', 2)
print(img)
ret, Binary_image = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
Binary_image = Binary_image[2100:2800,2000:4000]

plt.figure(figsize=(20,10))
plt.imshow(Binary_image, cmap=plt.get_cmap('gray'))
plt.axis('off')
#plt.savefig('binary.jpg', bbox_inches='tight')
plt.show()
```

### Converting RGB image to Indexed image

```python
from skimage import io
from sklearn.cluster import KMeans
```

```python
from skimage import io
from sklearn.cluster import KMeans

rgb = io.imread('truecolor.jpg')
n = 25 #number of indices/how many bins the colors will be placed in/number of colors

array = rgb.reshape((-1, 3))
kmeans = KMeans(n_clusters=n, random_state=42).fit(array)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
Indexed_image = centers[labels].reshape(rgb.shape).astype('uint8')

plt.figure(figsize=(20,10))
plt.imshow(Indexed_image)
plt.axis('off')
plt.savefig('indexed.jpg', bbox_inches='tight')
plt.show()
```
